```python
import os, sys

def get_root_path(search_word):
    """ Búsqueda hacia atrás de un directorio con nombre search_word. """
    cwd = os.getcwd()
    while str(os.getcwd()).split('/')[-1] != search_word:
        os.chdir("..")
        cwd = os.getcwd()
    return cwd

ROOT_DIR = get_root_path('TFM_StockForecast')

sys.path.insert(0, ROOT_DIR)
```

```python

df = load_data()
for prod in ["30","32","42","48"]:
    prod_data = df.loc[df.producto == prod]
    prod_data = prepare_train_data(prod_data)
    prod_data = prod_data[['fecha', 'producto', 'udsstock', 'udsventa', 'udsprevisionempresa',
       'promo', 'festivo', 'weekday', 
       'working_day', 'sin_weekday', 'cos_weekday', 'is_august', 'spring',
       'summer', 'autumn', 'winter','roll4wd_udsventa',
       'roll4wd_udsstock',
       'udsprevisionempresa_shifted-1', 'udsprevisionempresa_shifted-2',
       'udsstock_shifted1', 'udsstock_shifted7', 'roll4wd_udsstock_shifted1','roll4wd_udsstock_shifted7',
       'udsventa_shifted1']]
    prod_data["roll4_udstock_shifted1"] = prod_data['udsstock_shifted1'].rolling(4, win_type='triang', min_periods=1).mean()
    profile = ProfileReport(prod_data, title='EDA for STOCK dataset and product {}.'.format(prod))
    profile.to_file(output_file="reports/eda/EDA_Stock_{}_prepared.html".format(prod))
```
