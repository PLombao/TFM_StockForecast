### 1.1 Generación de datos

El siguiente código genera una serie temporal con cierta aleatoriedad, guarda la serie en la variable "serie" y la visualiza en una gráfica.

El periodo es completamente fundamental para calcular correctamente la descomposición.

+ Si ponemos un periodo igual a 1, la componente seasonal será cte e igual a 0, y por tanto la parte residual comprenderá tanto la estacionalidad como el resto.

+ Al aumentar el periodo desde 1 hasta 50, vamos capturando más estacionalidad en la parte seasonal, modulando la frecuencia y ensanchando los ciclos, pero siempre se nos escapa la mayor parte de la estacionalidad, quedano está reflejada en la componente Residual.

+ Cuando nos acercamos a 50 se van inviertiendo las tornas, y la parte Seasonal empieza a comportarse según lo esperado en las inmediaciones de 50.

+ Según nos alejamos vuelve a ocurrir lo mismo pero ahora con frecuencias cada vez más bajas, pero no se captura correctamente la seasonal.

+ En los múltiplos del periodo (100, 150, 200...) ocurre lo mismo que en el valor de la frecuencia.

### 1.3 Predicción de la tendencia

La tendencia de una serie temporal cambia lentamente con el tiempo. Además, en el momento de realizar predicciones normalmente toma valores fuera del rango que aparece en los datos de entrenamiento. Esto fuerza al modelo a extrapolar, por lo que hay que utilizar modelos muy robustos para predecir la tendencia.

<div style="background-color: #EDF7FF; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">
<strong>Ejercicio (0,5 ptos):</strong> Entrenar un modelo polinomial con las variables $x$ y $x^2$ para predecir los valores de la tendencia en el conjunto de entrenamiento. Dibujar una gráfica con la tendencia original y su predicción.
</div>

```python
# Realizamos el modelo usando PolynomialFeatures
from sklearn.preprocessing import PolynomialFeatures

x = idx[:, np.newaxis] # La variable dependiente será nuestro index
y = dec.trend # La variable independiente, nuestro target, será el trend de nuestra serie

polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x)

linear_model = LinearRegression()
linear_model.fit(x_poly, y)
y_pred = linear_model.predict(x_poly)


from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y,y_pred))
r2 = r2_score(y,y_pred)
print("RMSE for polynomial regression %.4f " % rmse)
print("R-square for polynomial regression %.4f " % r2)


plt.figure(figsize=(20,10))
plt.plot(idx, dec.trend, label='Actual')
plt.plot(idx, y_pred, label = 'Predicted')
plt.legend()
plt.xlabel('t')
plt.ylabel('Trend')
plt.title("Actual trend vs Predicted trend - Polynomial Model")
plt.show()
```

### 1.4 Predicción de la parte estacionaria

Al descomponer la serie temporal obtenemos la estacionalidad como una nueva serie temporal que se repite cada cierto tiempo, por lo que es muy adecuada para predecirla utilizando una red neuronal recurrente. Dado que las redes neuronales repiten varias veces la misma operación sobre un conjunto de datos, es importante normalizar los valores para evitar problemas de explosión o desaparición del gradiente.

<div style="background-color: #EDF7FF; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">
<strong>Ejercicio (0,5 ptos):</strong> Normalizar los valores de la serie estacionaria a valores entre 0 y 1.
</div>

```python
season = np.array(dec.seasonal).reshape(-1,1)

scaler = MinMaxScaler()

seasonal_norm = scaler.fit_transform(season)

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.plot(idx,dec.seasonal)
plt.title("Not normalized seasonal data")

plt.subplot(1,2,2)
plt.plot(idx, seasonal_norm)
plt.title("Normalized seasonal data")

plt.show()
```

El siguiente código genera el dataset de entrenamiento de la red recurrente. Utilizando lookback = 30 cada ejemplo del dataset contiene los 30 últimos valores de la serie y su etiqueta corresponde al siguiente valor.

```python
lookback = 30

features_set = [] 
labels = []  
for i in range(lookback, seasonal_norm.shape[0]):  
    features_set.append(seasonal_norm[i-lookback:i, 0])
    labels.append(seasonal_norm[i, 0])

features_set, labels = np.array(features_set), np.array(labels)
features_set = np.reshape(features_set, (features_set.shape[0], features_set.shape[1], 1))
```

<div style="background-color: #EDF7FF; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">
<strong>Ejercicio (0,5 ptos):</strong> Definir y entrenar una red neuronal recurrente con celdas GRU cuya salida esté formada por una capa completamente conectada de una única neurona con función de activación lineal.
</div>

```python
# Train test split, 80 - 20%
cut_index = int(labels.shape[0]*0.8)

features_train = features_set[:cut_index]
features_test = features_set[cut_index:]

labels_train = labels[:cut_index]
labels_test = labels[cut_index:]
```

```python
# Compilamos el modelo
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import GRU
from keras.models import Sequential
from keras import optimizers

model = Sequential()
model.add(GRU(50, return_sequences=True, input_shape=(features_set.shape[1],1)))
model.add(Dropout(0.2))
model.add(GRU(100, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1, activation = "linear"))

model.compile(loss='mse', optimizer='adam')
```

```python
# Entrenamos el modelo
model.fit(features_train, labels_train, epochs=10, 
          validation_data = (features_test, labels_test))
```

<div style="background-color: #EDF7FF; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">
<strong>Ejercicio (0,5 ptos):</strong> Utilizar el modelo entrenado anteriormente para predecir los valores del conjunto de entrenamiento de la serie estacionaria. Dibujar una gráfica con la tendencia original y su predicción.
</div>

```python
label_pred = model.predict(features_set)

rmse = np.sqrt(mean_squared_error(labels,label_pred))
r2 = r2_score(labels,label_pred)
print("RMSE for seasonal forecast %.4f " % rmse)
print("R-square for seasonal forecast %.4f " % r2)

plt.figure(figsize=(20,10))
plt.plot(idx[30:], labels, label='Actual')
plt.plot(idx[30:], label_pred, label = 'Predicted')
plt.legend()
plt.xlabel('t')
plt.ylabel('Seasonal')
plt.title("Actual vs Predicted Seasonal - RNN with GRU Model")
plt.show()
```

## 2. Clasificación de textos (5 puntos)

En este apartado trabajaremos con el conjunto de datos de noticias de Reuters clasificadas en 46 categorías diferentes. En lugar de clasificar la categoría para cada noticia, entrenaremos una red neuronal siamesa que prediga si dos noticias pertenecen a la misma categoría o no.

### 2.1 Carga y preprocesado de los datos

El siguiente código carga los datos de las noticias directamente del repositorio de Keras.

```python
print("Cargando datos...")
num_words = 10000
(x_train_orig, y_train_orig), (x_test_orig, y_test_orig) = keras.datasets.reuters.load_data(num_words=num_words)
print("Finalizado")
```

Cada noticia tiene una longitud diferente pero, para facilitar el entrenamiento de la red neuronal recurrente, igualaremos la longitud de todas las secuencias utilizando un carácter especial de "padding".

```python
x_train = keras.preprocessing.sequence.pad_sequences(x_train_orig, maxlen=300)
x_test = keras.preprocessing.sequence.pad_sequences(x_test_orig, maxlen=300)

y_train = y_train_orig
y_test = y_test_orig
```

### 2.2 Preparación de los conjuntos de entrenamiento y test

Para entrenar un modelo que prediga si dos noticias pertenecen o no a la misma categoría debemos crear conjuntos de datos aparejados. El siguiente código genera los conjuntos de entrenamiento y test, cada uno con 20.000 parejas escogidas al azar. Es importante que las parejas para cada conjunto se generen de forma que las dos noticias que forman una pareja pertenezcan al mismo conjunto originalmente (entrenamiento o test), de forma que en el conjunto de parejas de test no aparezca en ningún caso una noticia que también estuviera en el conjunto de entrenamiento.

```python
n_examples = 20000

train_pairs = np.random.choice(range(x_train.shape[0]), (n_examples, 2))
test_pairs = np.random.choice(range(x_test.shape[0]), (n_examples, 2))
```

```python
x_train_left = x_train[train_pairs][:,0]
x_train_right = x_train[train_pairs][:,1]

y_train_pairs = y_train[train_pairs]
train_labels = (y_train_pairs[:,0] == y_train_pairs[:,1]).astype(np.int)
```

```python
x_test_left = x_test[test_pairs][:,0]
x_test_right = x_test[test_pairs][:,1]

y_test_pairs = y_test[test_pairs]
test_labels = (y_test_pairs[:,0] == y_test_pairs[:,1]).astype(np.int)
```

### 2.3 Definición y entrenamiento de la red neuronal

A continuación utilizaremos una red recurrente siamesa para resolver el problema de clasificación que hemos preparado. Una red recurrente siamesa es una red neuronal que tiene una parte recurrente que se aplica a dos entradas diferentes para obtener una "codificación" (los mismos pesos se utilizan con las dos entradas por lo que no importa si, en una pareja, una noticia va en primera o en segunda posición). Entonces, la red siamesa contiene otra parte no recurrente que combina la codificación de las dos entradas para predecir una clasificacion.

El siguiente diagrama muestra la red neuronal recurrente que utilizaremos:
<img src="siamese_network.png" alt="Diagrama de la red neuronal siamesa" style="height: 500px;"/>

<div style="background-color: #EDF7FF; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">
<strong>Ejercicio (2 ptos):</strong> Definir y entrenar la red neuronal recurrente que se muestra en el diagrama.
</div>

```python
from keras.layers import Input, Embedding, LSTM, concatenate, Dense, Dropout
from keras.models import Model

# Preparamos las dos capas de entrada paralela
input_left = Input(x_train_left.shape[1:])
input_right = Input(x_train_right.shape[1:])

# Preparamos la capa de embedding y LSTM que se aplicaran paralalamente a los dos entradas
words_embedding_layer = Embedding(x_train_left.max()+1, 50)
seq_embedding_layer = LSTM(32, activation='tanh')

# Definimos una función lambda para aplicar embedding y LSTM a las dos entradas
embedding = lambda tensor: seq_embedding_layer(words_embedding_layer(tensor))

# Concatenamos las dos entradas
merge_layer = concatenate([embedding(input_left),embedding(input_right)])

# Creamos una capa de salida
pred_layer = Dense(20, activation='relu')(merge_layer)
#drop_layer = Dropout(0.2)(pred_layer)
output_layer = Dense(1, activation='sigmoid')(pred_layer)

# Creamos el modlo
model = Model([input_left, input_right], output_layer)

# Compilamos
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
train = model.fit([x_train_left, x_train_right], train_labels,
         validation_data=([x_test_left, x_test_right],test_labels),
         batch_size=128, epochs=6)
```

```python
## TODO: plot del training loss y el accuracy
def plot_prediction(n_epochs, mfit):
    plt.figure(figsize=(15,18))
    plt.style.use('ggplot')

    x = np.arange(1, n_epochs +1)
    # Ploteamos el primer subplot con la evolucion de la precision
    plt.subplot(2, 1, 1)
    plt.plot(x, mfit.history['acc'], label = 'train')
    plt.plot(x, mfit.history['val_acc'], label = 'test')
    plt.title('Accuracy Evolution')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.legend(loc='lower right', fontsize = 'x-large')

    plt.subplot(2, 1, 2)
    plt.plot(x, mfit.history['loss'], label = 'train')
    plt.plot(x, mfit.history['val_loss'], label = 'test')
    plt.title('Loss Evolution')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend(loc='upper right', fontsize = 'x-large')

    plt.tight_layout()
    plt.show()
    
    return plt
plot_prediction(6, train)
```

### 2.4 Análisis de los resultados

Dado que el problema de clasificación que hemos utilizado para entrenar la red neuronal tiene únicamente dos clases (0 y 1) una métrica muy útil para validar el rendimiento del modelo es el AUC ROC.

<div style="background-color: #EDF7FF; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">

</div>

```python
y_pred_train = model.predict([x_train_left,x_train_right])
y_pred_test = model.predict([x_test_left,x_test_right])
```

<strong>Ejercicio (1,5 ptos):</strong> Comentar la relación entre el crecimiento de "True Positive Rate" y el crecimiento de "False Positive Rate".
</div>

+ La **sensitivity** o recall (también conocida como TPR o **True Positive Rate**) **mide el ratio de positivos correctamente clasificados** (es decir, los positivos etiquetados como tal entre todos los positivos reales)

+ La **specificity** se trata del **ratio de negativos correctamente clasificados** (es decir, los negativos etiquetados como tal entre todos los negativos reales).

+ El FPR o **False Positive Rate** mide el **ratio de negativos erróneamente clasificados** (es decir, los negativos etiquetados mal entre todos los negativos reales). Es complementario de la especificidad, o lo que es lo mismo, **el FPR y la specificity suman 1**.

+ La curva ROC nos dice la relación entre el ratio de positivos clasificados bien y negativos clasificados mal según variamos el threshold para separar las clases, de manera que el area debajo de la curva o AUC, nos determina como de bien es capaz de separar el modelo las clases. Un modelo con un AUC de 0.5 quiere decir que no es capaz de separar correctamente las clases, mientras que un AUC próximo a 1 quiere decir que nuestro modelo ha aprendido correctamente a separar las clases.

+ Este valor de 1 se da cuando las distribuciones de TN y TP son separables totalmente por un determinado threshold, de esta manera existe un valor para el cual podemos separar perfectamente las dos clases y la curva ROC se comporta de manera ideal (una curva en ángulo recto con puntos en 0,0, 0,1 y 1,1).

+ Cuando las distribuciones se sobreponen ligeramente, introducimos errores de Tipo I y de Tipo II, es decir FP y FN y según nuestro threshold podemos modular para favorecer unos u otros. En estos casos la curva se comporta como la nuestra y con un ROC en test de 0.81 que quiere decir que **tenemos un 81% de probabilidades de que el modelo sepa distinguir entre las dos clases**.

+ En el caso en que las distribuciones se solapen, la curva ROC será similar a la línea discontinua dibujada arriba con un AUC cercano a 0.5.

+ El caso más extremo sería que el modelo estuviera confundiendo las distribuciones y etiquetando las clases al revés, en ese caso tendríamos una curva cóncava con un AUC cercano a 0.

+ El **TPR y la specificity son inversamente proporcionales** dado un modelo determinado. De manera que **si movemos el threshold para aumentar el TPR** (o reducir el número de falsos positivos), **disminuiremos la specificity** (aumentando el número de falsos negativos).

+ Dado que el FPR es 1 - specificty, al aumentar el TPR disminuimos el FPR y viceversa. Esto se puede entender más fácil si se mira por los errores: **si queremos aumentar el TPR tenemos que disminuir los Falsos Negativos, si disminuimos los Falsos Negativos entonces aumentan los Falsos Positivos y por tanto aumenta también el FPR**. Por tanto **el True Positive Rate y False Positive Rate son directamente proporcionales**.
