```python
import os, sys

def get_root_path(search_word):
    """ Búsqueda hacia atrás de un directorio con nombre search_word. """
    cwd = os.getcwd()
    while str(os.getcwd()).split('/')[-1] != search_word:
        os.chdir("..")
        cwd = os.getcwd()
    return cwd

ROOT_DIR = get_root_path('TFM_StockForecast')

sys.path.insert(0, ROOT_DIR) # Poner en el primer lugar del PATH (evitar paquetes con mismos nombres en otros directorios)
```

```python
import pandas as pd
import numpy as np
from src.load_data import load_clustering_data, load_data
from src.prepare_data import prepare_train_data
from src.trainer import data_producto
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_df = load_data()
```

# Features Stock

```python
from sklearn.ensemble import RandomForestRegressor
from src.trainer import run

def train(data, prod, variables):
    target="udsstock"
    data = data.loc[data.stockMissingType == 0].reset_index(drop=True)
    data = data.loc[data.producto == prod]
    data = prepare_train_data(data)
    data = data[variables]
    
    
    tags = {"type":"monoproducto",
            "base_model": "random_forest",
            "target":target}
    params = {"n_estimators":200}
    
    
    tags["name"] = "PR" + prod
        
    base_model = RandomForestRegressor(n_estimators=200)
        
    model, metrics, predict, cv_metrics = run(data, target, base_model, params, tags)
    
    return cv_metrics, predict
```

```python
features = {"udsstock_shifted1":["fecha", 'producto','udsprevisionempresa','udsventa_shifted1', "udsstock", "udsstock_shifted1"],
            "udsstock_shifted7" : ["fecha", 'producto','udsprevisionempresa','udsventa_shifted1', "udsstock", "udsstock_shifted7"],
            "roll4wd_udsstock_shifted1" : ["fecha", 'producto','udsprevisionempresa','udsventa_shifted1', "udsstock", "roll4wd_udsstock_shifted1"],
            "roll4wd_udsstock_shifted7" : ["fecha", 'producto','udsprevisionempresa','udsventa_shifted1', "udsstock", "roll4wd_udsstock_shifted7"],
           }

all_metrics = []
for feature in list(features):
    
    metrics, _ = train(main_df, "30", features[feature])
    metrics["model"] = feature
    all_metrics.append(metrics)
all_metrics
```

```python
met = pd.DataFrame(all_metrics).set_index("model")
met["acc"] = 100 - met["mape_mean"]
fig, axes = plt.subplots(nrows=1, ncols=2)
met[["mae_mean","rmse_mean"]].plot(figsize=(20,16),ax=axes[0], kind='bar', color=["#D65F5F", "grey"])
axes[0].set_ylabel("udsstock")
axes[0].set_xlabel("")
met[["acc"]].plot(figsize=(20,16),ax=axes[1], kind='bar', color=["#77BEDB"])
axes[1].set_ylabel("accuracy [%]")
axes[1].set_xlabel("")


plt.savefig("reports/features/stock_variants.png")
```
